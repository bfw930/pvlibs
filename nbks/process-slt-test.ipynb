```python
# imports
import numpy as np
from scipy.stats import linregress
from scipy.interpolate import splev, splrep


# initialise figure and axes
_w = 9; _h = 6
fig = plt.figure(figsize = (_w, _h))
fig.canvas.layout.width = '{}in'.format(_w)
fig.canvas.layout.height= '{}in'.format(_h)


# select node
node = db[0]

# unpack data
W = node['wafer_thickness']
o = node['wafer_optical_const']

time = node['time']
nd = node['nd']
tau = node['tau']

isuns = node['isuns']
ivocs = node['ivocs']

light = node['illumination']

# sort data
j = np.argsort(nd)

time = time[j]
nd = nd[j]
tau = tau[j]

ivocs = ivocs[j]
isuns = isuns[j]

light = light[j]



#spl = splrep( time, nd, )
#dnd = splev(time, spl, der = 1)

# get 1 sun voc
spl = splrep(isuns, ivocs)
voc = splev(1., spl, der = 0)
print('{:.3f} V'.format(voc))


# extrapolate suns-voc from log-linear regression
slope, icept, err1, err1, err3 = linregress( ivocs, np.log(isuns))
vs = np.arange(.1, .8, .01)
iss = np.exp(slope*vs + icept)


j = np.where(iss <= voc)
iss = iss[j]
vs = vs[j]

plt.plot(ivocs, -(isuns-1), '.-g')

I = -(iss-1)
P = I*vs
V = vs

plt.plot(V, I, '--r')
plt.plot(V, P, '-.b')

# rough maximum power point
k = np.where(P == P.max())
Vj = V[j][k]

# b-pline fit around maximum power point (rel. power)
k = np.where( (V[j] > Vj-0.1) & (V[j] < Vj+0.1) )
spl = splrep( V[j][k], P[k], )

# derivative of power at maximum power point
xr = np.arange(Vj-0.05, Vj+0.025, .001)
dP = splev(xr, spl, der = 1)

# linear regression for maximum power point voltage
slope, icept, r_value, p_value, std_err = linregress(dP, xr)
Vmpp = icept

# power at maximum power point
Pmpp = splev(Vmpp, spl, der = 0)

plt.plot(Vmpp, Pmpp, 'oy')


# b-pline fit around maximum power point (rel. power)
spl = splrep(V, I)
Impp = splev(Vmpp, spl, der = 0)

plt.plot(Vmpp, Impp, 'og')


FF = Impp*Vmpp/(1.*voc)
print(FF)


plt.xlabel('iVoc')
plt.ylabel('iSuns')

#plt.yscale('log')
plt.xlim(.2, voc*1.02)
plt.ylim(-.1, 1.1)

plt.tight_layout()
plt.show()
```

```python
# imports
import numpy as np
from scipy.stats import linregress
from scipy.interpolate import splev, splrep


# initialise figure and axes
_w = 9; _h = 6
fig = plt.figure(figsize = (_w, _h))
fig.canvas.layout.width = '{}in'.format(_w)
fig.canvas.layout.height= '{}in'.format(_h)


# select node
node = db[8]

isuns = node['isuns']
ivocs = node['ivocs']

# sort data
j = np.argsort(ivocs)
ivocs = ivocs[j]
isuns = isuns[j]

plt.plot(ivocs, -(isuns-1), '.-g')


# get 1 sun voc
spl = splrep(isuns, ivocs)
voc = splev(1., spl, der = 0)
print('iVoc {:.3f} V'.format(voc))


# extrapolate suns-voc from log-linear regression
slope, icept, err1, err1, err3 = linregress( ivocs, np.log(isuns))
V = np.arange(.1, .8, .01)
iss = np.exp(slope*V + icept)

I = -(iss-1)
P = I*V

j = np.where(I >= 0.)
I = I[j]
V = V[j]
P = P[j]


plt.plot(V, I, '--r')
plt.plot(V, P, '-.b')


# rough maximum power point
k = np.where(P == P.max())
Vj = V[j][k]

# b-pline fit around maximum power point (rel. power)
k = np.where( (V[j] > Vj-0.1) & (V[j] < Vj+0.1) )
spl = splrep( V[j][k], P[k], )

# derivative of power at maximum power point
xr = np.arange(Vj-0.05, Vj+0.025, .001)
dP = splev(xr, spl, der = 1)

# linear regression for maximum power point voltage
slope, icept, r_value, p_value, std_err = linregress(dP, xr)
Vmpp = icept

# power at maximum power point
Pmpp = splev(Vmpp, spl, der = 0)

plt.plot(Vmpp, Pmpp, 'ob')


# b-pline fit around maximum power point (rel. power)
spl = splrep(V, I)
Impp = splev(Vmpp, spl, der = 0)

plt.plot(Vmpp, Impp, 'or')


pFF = Impp*Vmpp/(1.*voc)
print('pFF {:.3f}'.format(pFF))


plt.xlabel('iVoc')
plt.ylabel('iSuns')

#plt.yscale('log')
plt.xlim(.2, voc*1.02)
plt.ylim(0., 1.1)

plt.tight_layout()
plt.show()
```
